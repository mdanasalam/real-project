```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Car Price Analysis using Excel.xlsx")
```

```python
df
```

```python
df = df.dropna(subset=["year"])
```

```python
df
```

```python
year_price = df.groupby("year")["selingprice"].mean()

plt.figure(figsize=(8, 4))
plt.plot(year_price.index, year_price.values, marker="o")
plt.xlabel("Year")
plt.ylabel("Average selingprice")
plt.title("Average selingprice by Year")
plt.grid(True)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(df["odometer"], df["selingprice"], alpha=0.7)
plt.xlabel("Odometer Reading")
plt.ylabel("selingprice")
plt.title("Odometer vs selingprice")
plt.grid(True)
```

```python
make_price = (
    df.groupby("make")["selingprice"]
      .mean()
      .sort_values(ascending=False)
      .head(5)
)

plt.figure(figsize=(8, 4))
plt.bar(make_price.index, make_price.values, color="skyblue")
plt.xlabel("Make (Brand)")
plt.ylabel("Average selingprice")
plt.title("Top 5 Makes by Average selingprice")
plt.xticks(rotation=45)
plt.tight_layout()
```
